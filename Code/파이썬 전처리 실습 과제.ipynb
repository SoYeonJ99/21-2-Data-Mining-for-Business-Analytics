 ```python
 dataframe.drop(columns=열인덱싱 or 네이밍 등등)
 ```
 
 
3. 중복제거
 - 중복을 제거하는 함수: drop_duplicates(subset=['중복을 제거할 열'])
 - market, operator에는 merge가 포함된 열이 있음. 제거하기
 - https://yganalyst.github.io/data_handling/memo_9/



4. missing data를 median으로 대체하기
 - 참고: imputer는 문자형 변수가 있으면 안 돌아감.

````python
#1 데이터 파일 불러오기
xls = pd.ExcelFile("D:/숙탯 2기/4주차_데이터 전처리/RawData.xlsx")
raw_df = xls.parse('Raw Aggregate',na_values=['NA'])
````

````python
raw_df
````

````python
#2 의미없는 column 제거


#의미있는 column은 65개
raw_df.filter(regex = 'Q')    # Market~ 열 + Q~ 열 = 65개
````

````python
raw_df = raw_df.drop(columns = raw_df.columns[66:])
````

````python
raw_df
````

````python
#3 중복 제거

raw_df = raw_df.drop_duplicates(subset=raw_df[raw_df['Market, Operator'].str.contains("merge")])
raw_df
````

````python
#4 결측값 중간값 대체


#결측치 확인
print(raw_df.isna().sum())
````

````python
raw_df.info()     #1열만 object형
````

````python
from numpy import nan
from sklearn.impute import SimpleImputer
values = raw_df.iloc[:, 1:].values       #object형인 1열 빼고 실행
imputer = SimpleImputer(missing_values = nan, strategy='median')
medianimputed = pd.DataFrame(imputer.fit_transform(values))
medianimputed
````
